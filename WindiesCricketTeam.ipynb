# Windies Cricket Team Case Study


### In this case study I will point out the downfall and What went wrong for West Indies Cricket. Here I have collected the Dataset from the Years 2016 to 2023. Windies last won the world in the year 2016 after that they played in two world cups in the year 2021 and 2022. In the 2021 World Cup, They won only one match against Bangladesh in their group stage. They even failed to directly qualify for the group and went on to play qualifiers where they faced embarrassing losses and won only one match against  Zimbabwe. Not able to qualify for the group stage. Here I have collected the stats related to their series records, cumulative records, batsmen and bowler. I am going to use Python with pandas and power bi for visualization.

```python

import pandas as pd

df = pd.read_html("https://stats.espncricinfo.com/ci/engine/team/4.html?class=3;spanmin1=16+Feb+2016;spanval1=span;template=results;type=team;view=cumulative")[3]
```

```python
df.columns
```

```python
x = ['Tied', 'NR', 'W/L', 'Ave', 'RPO', 'Unnamed: 11','Start Date', 'Unnamed: 14']
```

```python
df.drop(x,axis=1,inplace=True)
```

```python
df1.head()
```

```python
df1.columns
```

```python
df1.head(10)
```

```python
for i in df1['Season']:
    print(i)
```

```python
x = ['2015/16',
'2016/17','2017','2017','2018',
'2018','2018/19','2019','2019/20','2019/20','2020/21','2021','2021/22','2021/22','2022','2022','2022/23']
df1 = df1[df1['Season'].str.contains('|'.join(x))]
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
df1['Winner'].describe()
```

```python
windieswinners = df1[df1['Winner'] == 'West Indies']
drawn_count = df1[df1['Winner'] == 'drawn']['Winner'].count()

sum_per_season = df1.groupby('Season')['Winner'].count()

print("West Indies is the winner:")
print(windieswinners)

```

```python
windieswinners.reset_index(drop=True, inplace=True)
windieswinners['Winner'].value_counts()
```

```python
windiesnotwinners = df1[df1['Winner'] != 'West Indies']
drawn_count = df1[df1['Winner'] == 'drawn']['Winner'].count()

sum_per_season = df1.groupby('Season')['Winner'].count()

print("West Indies is not the winner:")
print(windiesnotwinners)

```

```python
windiesnotwinners.reset_index(drop=True, inplace=True)
windiesnotwinners['Winner'].value_counts().sum()
```
